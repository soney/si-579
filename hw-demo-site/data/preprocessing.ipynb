```python
import regex as re
import os
import pandas as pd
```

```python
namebook = pd.read_csv('./namebook.csv')
```

```python
namebook.head()
```

```python
def convertName(x):
    x['First Name'] = x.Student.split(', ')[1]
    x['Last Name'] = x.Student.split(', ')[0]
    x['UID'] = x['Last Name'] + x['First Name'].replace(' ', '').replace('-','')
    x['UID'] = x['UID'].lower()
    x['Opt-out'] = False
    return x
```

```python
namebook = namebook.apply(convertName, axis = 1)
```

```python
namebook.head()
```

```python
pid = 1

for pid in range(1, 5):
    pname = 'PS' + str(pid)
    basepath = 'data/' + pname
    entries = os.listdir(basepath)
    for entry in entries:
        f = open(os.path.join(basepath, entry), "r")
        content = f.read()
        match = re.search('url=(.*)">', content)
        url = match.group(1)
        match = re.search('(.*)_(.*)_', entry)
        uid = match.group(1).split('_')[0]
        index = namebook.index[namebook['UID'] == uid].tolist()
        if(len(index)==0):
            print(uid)
        else:
            namebook.loc[index[0], pname] = url
```

```python
namebook.to_json('data.json', orient = "records")
```
